Library 

```python
import pandas as pd
import seaborn as sns
```

Convert dataset from excel into Pandas dataframe

```python
df = pd.read_excel("https://github.com/Prenagen24/Life-Expectancy-on-All-Continent-based-on-GDP-and-Population/blob/main/dataset_tugas2.xlsx?raw=true")
df

# we renaming the "life Expectancy at Birth" to "LEB" 
```

Slice selected column from the main dataframe into new dataframe

```python
df_new = df[['Continent','GDP','LEB','Population']]
df_new
```

Visualization using Seaborn

```python
sns.set_theme()
sns.set(rc={"figure.figsize":(24, 16)})
cmap = sns.color_palette("Set2",n_colors=5)
g = sns.relplot(
    data=df_new,
    x="Population", y="GDP",
    hue="Continent", size="LEB",
    palette=cmap,sizes=(10,200)
)
g.set(xscale="log", yscale="log",title='4 Dimensions plot between GDP, Population, LEB, and Continent ')
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
